```python
files = ['chinese_food.json', 'indian_food.json', 'japanese_food.json', 'muslim_food.json', 'western_food.json']
result_files = ['clean_' + text for text in files]
result_files_formated = ['formatted' + text for text in result_files]
```

```python
def sanitizerJSON(path_to_file, path_to_result, selection=None):

    import json 
    with open(path_to_file, encoding='utf-8') as file:
        f = json.load(file)
        print("Keys:", f.keys())
        data = f.get('data')
        data_1 = data[0]

        # Set Up
        if selection:
            result = {}
            for index, d in enumerate(data):
                    result.update({index: {}})
                    for k in selection:
                        result.get(index).update({k: data[index].get(k)})
        else: 
            dict_of_keys = {}

            for index, key in enumerate(data_1.keys()):
                dict_of_keys.update({index: key})

            # Get Information on Structure of data
            print("\nLength of keys:", len(data_1.keys()))
            print("\n".join([" - ".join((str(i), j )) for i, j in enumerate(data_1.keys())]))

            # Select Data to Show
            data_to_extract = [int(i) for i in input("Enter index of data you want").split(',')]

            # Extract Data
            result = {}
            for index, d in enumerate(data):
                result.update({index: {}})
                for k in data_to_extract:
                    key = dict_of_keys.get(k)
                    result.get(index).update({key: data[index].get(key)})

        # Write to JSON
        with open(path_to_result, 'w', encoding='utf-8') as result_file:
            r = json.dumps(result)
            result_file.write(r)
        return result

for i in range(len(files)):
    sanitizerJSON(files[i], result_files[i], ['thumbnails', 'images', 'body', 'uuid', 'rating', 'contact', 'officialWebsite', 'officialEmail', 'description', 'address', 'name'])
```

```python
import json
for i in range(5):
    with open(result_files[i], encoding='utf-8') as file:
        f = json.load(file)
        result = {}
        for key in f.keys():
            data = f[key]
            result.update({key: {}})
            for k in data.keys():
                if k == 'contact':
                    result[key].update({'primaryContactNo': data[k].get('primaryContactNo')})
                    result[key].update({'secondaryContactNo': data[k].get('secondaryContactNo')})
                elif k == 'address':
                    result[key].update({'block': data[k].get('block')})
                    result[key].update({'streetName': data[k].get('streetName')})
                    result[key].update({'floorNumber': data[k].get('floorNumber')})
                    result[key].update({'unitNumber': data[k].get('unitNumber')})
                    result[key].update({'buildingName': data[k].get('buildingName')})
                    result[key].update({'postalCode': data[k].get('postalCode')})
                elif k == 'images':
                    length = len(data['images'])
                    if length == 0:
                        result[key].update({'image_one': ""})
                        result[key].update({'image_two': ""})
                    elif length == 1:
                        result[key].update({'image_one': data['images'][0].get('url')})
                        result[key].update({'image_two': ""})
                    else:
                        result[key].update({'image_one': data['images'][0].get('url')})
                        result[key].update({'image_two': data['images'][1].get('url')})
                elif k == 'thumbnails':
                    length = len(data['thumbnails'])
                    if length == 0:
                          result[key].update({'thumbnail': ""})
                    else:
                        result[key].update({'thumbnail': data['thumbnails'][0].get('url')})
                else:
                    result[key].update({k: data[k]})
            with open(result_files_formated[i], 'w', encoding='utf-8') as result_file:
                r = json.dumps(result)
                result_file.write(r)
```
